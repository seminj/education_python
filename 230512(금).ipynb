```python
# 1번 방법 (나)
def func9(text):
    Charn = "!@#$%^&*()_+[]{}  !#$%^&*()@{}_+[]" 
    for Char in Charn:
        text = text.replace(Char, '')
    print(text.replace(' ', '').replace('\n', ''))
```

어떠한 수 N이 될 때까지 다음의 두 과정 중 하나를 반복적으로 선택하여 수행하려고 합니다. 단, 두번째 연산은 N이 K로 나누어 떨어질 때만 선택할 수 있습니다. </br>
    1. N에서 1을 뺍니다. / 2. N을 K로 나눕니다. </br>
예를 들어 N이 17, K가 4라고 가정합시다. 이 때 1번의 과정을 한 번 수행하면 N은 16이 됩니다. 이후에 2번의 과정을 두 번 수행하면 N은 1이 됩니다. 결과적으로 이 경우 전체과정을 실행한 횟수는 3이 됩니다. 이는 N을 1로 만드는 최소 횟수입니다. N과 K가 주어질 때 N이 1이 될 때까지 1번 혹은 2번의 과정을 수행해야 하는 최소 횟수를 구하는 프로그램을 작성하세요.

```python
# 1번 방법
N = 26
K = 3
count = 0
loop = 0

# while N >1:
while True:
    loop += 1
    if N % K == 0:
        N //= K 
    else:
        N -= 1
    count += 1
    if N ==1: 
        break
print(count, loop)
```

```python
# 2번 방법
N = 26
K = 3
count = 0
loop = 0

while True:
    loop += 1
    if N % K == 0:
        N //= K
        count += 1
    else:
        나머지 = N % K
        count += 나머지
        N -= 나머지
    if N == 0: 
        count -= 1
        break # N이 0이 되었을 때 break
print(count, loop) # 몇번을 나눌지 모르니까 나머지로 계산
```

#### lambda : 단순한 반복문만 사용가능

```python
def plus(a, b):
    return a + b
result = plus(10, 20)
print(result)
```

```python
plus = lambda a, b : a + b
result = plus(10,20)
print(result)
```

```python
# a ~ z 문자열 생성하기
# 1. 리스트 컴프리헨션
a = [ chr(i) for i in range(ord('a'), ord('z') + 1) ]
print(a)
# print(''.join(a)) # 붙여쓰기
```

```python
# 2. lambda, map
a = [ i for i in range(ord('a'), ord('z') + 1) ]
# print(a)
b = list(map(lambda x : chr(x), a))# 값이 두개 들어감
''.join(b)
```

#### 연습문제 

```python
# 조건문 : [ [ 표현식 for 항목 in 반복가능 객체 if 조건] ]
```

```python
# 아래의 함수를 lambda와 List Comprehension을 사용하는 함수로 작성
def myfunc(numbers):
    result = []
    for number in numbers:
        if number > 5:
            result.append(number)
    return result

result = myfunc([2, 3, 4, 5, 6, 7, 8])
print(result)
```

```python
items = [2, 3, 4, 5, 6, 7, 8]
# [i for i in items if i > 5 ]
myfunc = lambda numbers: [i for i in numbers if i > 5]
myfunc(items)
```

#### 파일쓰기

write() - 문자를 파일의 내용으로 쓰기

writelines() - 리스트를 파일의 내용으로 쓰기

### 날씨 파일

#### 1. 04_seoul_기온.csv 파일 불러오기

```python
f = open('04_seoul_기온.csv')
data = f.readlines()
f.close()
```

```python
data
```

#### 2. 날짜와 최저기온 데이터 정리

```python
# date_list와 min_list 생성
date_list = []
min_list = []
max_list = []
avg_list = []

# 반복문을 사용하여 각 데이터 저장
for d in data[1:]: # 첫번째꺼는 지우고 데려온다
    date_list.append(d.split(',')[0])
    min_list.append(float(d.split(',')[-2]))
    max_list.append(float(d.split(',')[-1]))
    avg_list.append(float(d.split(',')[1]))
```

```python
# date_list
# min_list
# max_list
avg_list
```

#### 3. 시각화 준비 (환경설정) / 4. 그래프로 데이터 표현하기

```python
# 경고창 없애기
import warnings
warnings.filterwarnings(action='ignore') 

import matplotlib.pyplot as plt # matplotlib.pyplot 모듈 가져오기
# Line 그래프 사용 (plot)
plt.xticks(range(len(min_list)), date_list, rotation = 45)
plt.plot(max_list)
plt.plot(min_list) 
plt.plot(avg_list)
```
