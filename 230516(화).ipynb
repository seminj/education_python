## 최대 공약수

#### GCD : Greatest Common Divisor

```python
# 21, 14 => 7
n1 = 21
n2 = 14 # 작은 값에서 -1씩해야 숫자가 나오니까
for i in range(min(n1, n2), 0, -1):
    if n1 % i == 0 and n2 % i == 0:
        print(f'최대 공약수는 {i}')
        break
```

#### 유클리드 호제법Euclidean agorithm

- x와 y가 있고 x를 y로 나눈 나머지 r이 있을 때
- x와 y의 최대 공약수는 y와 r의 최대 공약수와 같다

```python
x = 16
y = 14
r = x % y
print(f'x: {x}, y: {y}, r: {r}')

x = y
y = r
r = x % y
print(f'x: {x}, y: {y}, r: {r}')

print(f'r의 값이 0이 되었으므로 x 16과 y 14의 최대 공약수는 {y}')
```

```python
def gcd(x, y):
    if y == 0:
        return x
    else:
        return gcd(y, x % y)

x = 16
y = 14

print(f'두 정숫값의 최대 공약수는 {gcd(x, y)}입니다.')
```

```python
def gcd(x, y):
    while True:
        r = x % y
        if r == 0: break
        x, y = y, r # 값을 바꾸는 것 
    return y
x = 16
y = 14

print(f'두 정숫값의 최대 공약수는 {gcd(x, y)}입니다.')
```

## 최소 공배수

#### LCD : Least Common Multiple

```python
# 21, 14 => 7
n1 = 21
n2 = 14
for i in range(max(n1, n2), n1 * n2 + 1):# max 큰값에서 시작
    if i % n1 == 0 and i % n2 == 0:
        print(f'최소 공배수는 {i}')
        break
```

#### 유클리드 호제법Euclidean agorithm

- x와 y가 있고 x * y를 최대 공약수로 나눈 몫은 최소 공배수이다

```python
x = 16
y = 14

def lcm(x, y):
    result = (x * y) // gcd(x, y)
    return result

print(f'두 정숫값의 최소 공배수는 {lcm(x, y)}입니다.')
```

```python
# random.shuffle() : List 자료형의 요소를 무작위로 정렬
def shuffle(items):
    size = len(items)
    for i in range(size):
        num = random.randint(0, 4)
        items[0], items[num] = items[num], items[0]
    return items

shuffle([1, 2, 3, 4, 5])
```

```python
import re
pattern = re.compile('\d{6}-[1-4]{1}[0-9]{6}$')
text = '791111-5234567'
print(text, end=' <- ')

result = pattern.search(text)
if result:
    print('정상')
else:
    print('올바르지 못한 주민등록번호')
```

#### 연습문제 : 은행 계좌번호 형식 검사하기
- 조건 1) 숫자 3자리-숫자2자리-숫자6자리
- 조건 2) 첫번째 3자리 숫자의 시작은 반드시 1 이어야 한다.
- 조건 3) 두번째 2자리 숫자의 시작은 반드시 1 이어야 한다.

```python
import re
pattern = re.compile('1\d{3}-1\d{2}-\d{6}')
text = '123-12-123456'
print(text, end=' <- ')
result = pattern.search(text)
if result:
    print('확인되었습니다.')
else:
    print('계좌번호 다시 확인해주세요.')
```

#### 연습문제 :  문자열이 형식에 맞는지 검사하기
- 조건 1) '안녕하세요' 문자로 시작
- 조건 2) 내용 중 '열심히' 문자 포함
- 조건 3) '감사합니다.' 문자로 종료
* 힌트) 줄바꿈 문자(\n)는 . 에 해당되지 않음

```python
import re
pattern = re.compile('^안녕하세요(.*)열심히(.*)감사합니다[.]$')
text = '''
안녕하세요. 저는 OOO에서 태어났고...
열심히 하겠습니다. 감사합니다.
'''
# 코드 작성
text = text.replace('\n', '')
result = pattern.search(text)
print(pattern.search(text))
print(result.group(1))
```

#### 정규 표현식
1. 검사
2. 추출(뽑아내기)
3. 치환(변경하기)

#### () : 정규식 그룹화 (index)

```python
import re
pattern = re.compile('(\d{3})-(\d{4})-(\d{4})')
result = pattern.search('010-8478-8181')
print(result.group())
print(result.group(0))
print(result.group(1))
print(result.group(2))
print(result.group(3))
```

#### 정규식 그룹화 (name)

```python
import re
pattern = re.compile('(?P<a>\d{3})-(?P<b>\d{4})-(?P<c>\d{4})')
result = pattern.search('010-8478-8181')
print(result.groupdict()) # Dictionary 형태로 접근 가능
print(result.group())
print(result.group(0))
print(result.group('a')) # result.group(1)
print(result.group('b')) # result.group(2)
print(result.group('c')) # result.group(3)
```

#### 연습문제 : 문자열의 내용 중 이름과 전화번호 찾아내기

```python
import re
pattern = re.compile('이름[ ]*\s*:\s*(\w+),\s*전화번호\s*:\s*(\d+), 이메일 : (.+)') # 공백에 \s* or [ ]* 사용가능
text = '''
사용자 정보, 이름: 꼬렙, 전화번호 : 1234, 이메일 : seorab@naver.com
'''
# 코드 작성
result = pattern.search(text)
print(result)
print(result.group(1))
print(result.group(2))
print(result.group(3))
```

### findall()

#### 정규식과 매치되는 모든 문자열을 List로 반환

#### 모든 숫자 List로 반환

```python
pattern = re.compile('[^-]') # 하이픈(-)이 아닌 1개 문자
result = pattern.findall('010-8478-8181')
print(result)
pattern = re.compile('[^-]+') # 하이픈(-)이 아닌 1개 이상의 문자
result = pattern.findall('010-8478-8181')
print(result)
```

```python
import re
pattern = re.compile(':\s*([가-힣0-9a-z@.]+)') # 공백에 \s* or [ ]* 사용가능
text = '''
사용자 정보, 이름: 꼬렙, 전화번호 : 1234, 이메일 : seorab@naver.com
'''
# 코드 작성
result = pattern.findall(text)
print(result)
```

#### 에러 번호 찾아내기

```python
import re
pattern = re.compile('에러\s*(\d+)\s*:\s*([가-힣 ]*)') # 숫자, 메시지, 숫자, 메시지로 변환됨
result = pattern.findall('에러 404 : 페이지 없음\n 에러 500 : 서버 오류') # findall : 여러개 나온다 
# result = pattern.search('에러 404 : 페이지 없음\n 에러 500 : 서버 오류') # search : N개만 나옴
print(result) # 소괄호 안에 쓰면 그 것만 뽑아냄
```
