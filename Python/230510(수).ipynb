```python
# 3. player가 이길 때까지 반복해서 게임이 진행되어야 한다. 
import random
count = 0

while True:
    player = int(input("가위 0, 바위 1, 보 2 중에 입력하세요. :"))
    com = random.randint(0,2) 
    count += 1 # 게임진행횟수

    if (player + 1) % 3 == com:
        print('com 승')
    elif player == com :
        print('비겼음')
    else :
        print('player 승')
        break
    print(f'com: {com}, player: {player}')
print(f'게임 진행횟수: {count}')
```

#### 중복되지 않는 1~45 사이의 번호 6개 생성하기 -> 1등 당첨 도전

```python
import random

당첨번호 = [6, 11, 16, 19, 21, 32]
로또번호 = list()
시도횟수 = 0

while True:
    시도횟수 += 1
    # 100만번마다 한번씩 시도횟수 출력
    if 시도횟수 % 1000000 == 0:
        print(f'현재 시도횟수: {시도횟수}')
    로또번호 = set()
    while True:
        로또번호.add(random.randint(1, 45))
        if len(로또번호) == 6:
            break
    로또번호 = list(로또번호)
    로또번호.sort()
    if 로또번호 == 당첨번호: break
print(시도횟수)
```

#### 그리디 : 거스름 돈 

당신은 음식점의 계산을 도와주는 점원입니다. 카운터에는 거스름 돈으로 사용할 500원,
100원, 50원, 10원 짜리 동전이 무한히 존재한다고 가정합니다. 손님에게 거슬러 주어야
할 돈이 N원일 때 거슬러 주어야 할 동전의 최소 개수를 구하세요. 단, 거슬러 줘야 할 돈
N은 항상 10의 배수입니다.

```python
# 큰 액수부터
거스름돈 = 1260
동전 = 0

while True:
    if 거스름돈 - 500 >= 0:
        동전 += 1
        거스름돈 -= 500
    elif 거스름돈 - 100 >= 0:
        동전 += 1
        거스름돈 -= 100
    elif 거스름돈 - 50 >= 0:
        동전 += 1
        거스름돈 -= 50
    elif 거스름돈 - 10 >= 0:
        동전 += 1
        거스름돈 -= 10
        
    if 거스름돈 == 0: break
print(동전)
```

```python
거스름돈 = 1260
동전 = 0

while True:
    if 거스름돈 - 500 >= 0:
        동전 += 거스름돈 // 500
        거스름돈 %= 500 # 거스름돈 = 거스름돈 % 500
    elif 거스름돈 - 100 >= 0:
        동전 += 거스름돈 // 100
        거스름돈 %= 100 
    elif 거스름돈 - 50 >= 0:
        동전 += 거스름돈 // 50
        거스름돈 %= 50 
    elif 거스름돈 - 10 >= 0:
        동전 += 거스름돈 // 10
        거스름돈 %= 10 
        
    if 거스름돈 == 0: break
print(동전)
```

#### List Comprehension : 리스트를 편하게 만들어주는 방식

- [ [ 표현식 for 항목 in 반복가능 객체 if 조건] ]
- [ [ 표현식 1 if 조건 else 표현식2 for 항목 in 반복가능 객체] ]

```python
# List [1, 2, 3, 4]의 각 요소에 10을 곱하여 새로 List 생성

a = [1, 2, 3, 4]
result = []
for n in a:
    result += [n * 3] # result.append(n * 3)
print(result)
```

```python
a = [1, 2, 3, 4]
result = [n * 3 for n in a]
print(result)
```

```python
# 조건문 추가 가능
a = [1, 2, 3, 4]
result = [n * 3 for n in a if n % 2 == 0]
print(result)
```
