#### 중첩(2중) 반복문

시계에서 '시'와 '분'이 같은 모든</br>
단, '시'는 24시간으로 표기</br>
00시 00분 /01시 01분 / 23시 23분

```python
# 3중 반복문을 시험으로 낼까
for h in range(0, 24):
    for m in range(0, 60):
        if h == m:
            print(f'{h}시 {m}분')
```

```python
# 5시 5분이 한자리 수 05시 05분으로
for h in range(0, 24):
    for m in range(0, 60):
        if h == m:
            if (h // 10) == 1 or (h // 10) == 2:
                    print(f'{h}시 {m}분')
            else:
                print(f'0{h}시 0{m}분')
```

#### 버블 정렬

```python
# 반복 1번
# [5, 3, 2, 4, 1]
# [3, 5, 2, 4, 1]
# [3, 2, 5, 4, 1] ...
# [3, 2, 4, 1, 5]
```

```python
items = [5, 3, 2, 4, 1]
for j in range(len(items)):
    for i in range(len(items) -1):
        a = items[i] # 5
        b = items[i + 1] # 3
        if a > b :
            a, b = b, a # a 와 b 의 위치 교환
            items[i] = a
            items[i + 1] = b
    print(items)
```

#### def return 함수

```python
items = [5, 3, 2, 4, 1]
def 정렬(a, b):
    for j in range(len(items)):
        for i in range(len(items) -1):
            a = items[i] # 5
            b = items[i + 1] # 3
            if a > b :
                a, b = b, a # a 와 b 의 위치 교환
                items[i] = a
                items[i + 1] = b
        return items
#         print(items) # return items 대신에 사용하면 None이 됨
# 정렬(1,2)
result = 정렬(1,2)
print(result)
```

#### 홀수 마방진

- 규칙 1) 다음 숫자는 오른쪽 상단으로 이동
- 규칙 2) 범위(위 또는 오른쪽)를 벗어나면 반대쪽(아래 또는 왼쪽)으로 이동
- 규칙 3) 이동하려는 위치에 0이 아닌 값이 있으면 이동 전 위치의 아래로 이동

![](./0511_홀수마방진.jpg)

```python
m = [[0 for col in range(5)] for row in range(5)]
row = 0
col = 2
m[row][col] = 1 # (0,2) 위치에 1입력

for i in range(2, 26): #2~25
    x = row
    y = col
    row -=1 # 행 -1
    col +=1 # 열 +1
    
    if row < 0 :
        row = 4
    if col > 4:
        col = 0
    if m[row][col] != 0:
        row = x + 1
        col = y 
    
    m[row][col] = i    
```

```python
# 출력
for row in range(len(m)):
    for col in m[row]:
        print(col, end = '  ')
    print()
```
