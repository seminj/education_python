#### 연습문제
- 데이터베이스를 활용하는 코드에 예외처리 적용하기
1. SQL 표현이 잘못된 경우 예외처리
2. 어떤 경우로 종료가 되더라도 close()가 수행되도록 finally 처리

```python
import sqlite3
conn = sqlite3.connect('geoje.db')
cursor = conn.cursor()

sql = '''
SELECT id, name, desc, mapinfo, latitude, longitude
FROM tour
''' # FROM tour

try:
    cursor.execute(sql)
    result = cursor.fetchmany(10)
    for row in result:
        print(row[1], row[4], row[5])
except:
    print('오류!')    

finally:
    cursor.close()
    conn.close()
```

#### 연습문제: 사용자 인증 기능 작성하기
- 조건 1) NotExistIdError, WrongPasswordError 클래스 작성
- 조건 2) 오류 메시지를 전달할 수 있도록 생성자 작성
- 조건 3) id/pw가 blue/1234가 아닌 경우
- 'ID를 확인해주세요', 'PW를 확인해주세요' 메시지 출력

```python
class NotExistIdError( ① ):
# 생성자 작성 ( ② )
class WrongPasswordError( ③ ):
# 생성자 작성 ( ④ )
def login(id, pw):
# id 비교 조건문 작성 ( ⑤ )
# 조건에 따라 예외 발생 ( ⑥ )
# pw 비교 조건문 작성 ( ⑦ )
# 조건에 따라 예외 발생 ( ⑧ )
print('로그인 되었습니다.')
login('blue', 1234)
```

```python
class NotExistIdError(Exception):
    def __init__(self, message):
        super().__init__(msg)

class WrongPasswordError(Exception):
    def __init__(self, message):
        super().__init__(msg)

def login(id, pw):
    if id != 'blue':
        raise NotExistIdError('ID를 확인해주세요')
    if pw != 1234:
        raise WrongPasswordError('PW를 확인해주세요')
    print('로그인 되었습니다.')

try:
    login('blue', 1234)
except NotExistIdError as e:
    print(e.message)
except WrongPasswordError as e:
    print(e.message)
```

### 수업 외 CSS 선택자 정리
:nth-child(N)= 부모안에 모든 요소 중 N번째 요소</br>
A:nth-of-type(N)= 부모안에 A라는 요소 중 N번째 요소</br>
:first-child= 부모안에 모든 요소 중 첫번째 요소</br>
:last-child= 부모안에 모든 요소 중 마지막 요소</br>
A:first-of-type= 부모안에 A라는 요소 중 첫번째 요소</br>
A:last-of-type= 부모안에 A라는 요소 중 마지막 요소

#### nth-child
:nth-child(1) = 첫번째 선택</br>
:nth-child(2n) = 두번째 마다 선택</br>
:nth-child(2n+1) = 첫번째 요소부터 2번째 마다 선택</br>
:nth-child(2n+5) = 다섯번째 부터 2개 마다 선택</br>
:nth-child(-2n+5) = 첫번째 요소부터 5번째 요소까지 2개 마다 선택</br>
:nth-child(n+5) = 5번째 부터 모두 선택</br>
:nth-child(-n+5) = 앞에서부터 5개만 선택</br>
:nth-child(n+3):nth-child(-n+8) = 선택자를 두번 사용하여 교집합인 요소만 선택할 수 있다.</br>
:nth-child(odd) = 홀수 선택</br>
:nth-child(even) = 짝수 선택</br>

#### nth-of-type
p:nth-of-type(1) = 부모안에 요소 중 첫번째 p요소 선택</br>
p:nth-of-type(2n+1) = 부모안에 p요소 중 첫번째부터 2번째마다 선택</br>
p:nth-of-type(n+2):nth-of-type(-n+5) = 선택자를 두번 사용하여 교집합인 요소만 선택할 수 있다.</br>
※ :nth-child와 마찬가지로 모든 선택자 속성을 사용할 수 있다.</br>
※ :nth-child와 차이점은 모든 속성이 아닌 선택한 요소 중에서 선택을 한다.</br>

**first-child**</br>
:first-child = 부모안에 모든 요소 중 첫번째 요소 선택</br>
**last-child**</br>
:last-child = 부모안에 모든 요소 중 마지막 요소 선택</br>
**first-of-type**</br>
p:first-of-type = 부모안에 모든 p요소 중 첫번째 p요소 선택</br>
**last-of-type**</br>
span:last-of-type = 부모안에 모든 span요소 중 마지막 span요소 선택</br>

#### 티오베의 아이콘 다운로드
1. tiobe 폴더 생성
2. 아이콘 다운로드 후 tiobe 폴더에 저장

```python
img_list = soup.select('.td-top20 >img')
import os
if not os.path.isdir('tiobe'):
    os.mkdir('tiobe')

for img in img_list:
    file_name = f'http://tiobe.com{img.attrs.get("src")}'# 싱기싱기
    print(file_name)
    res = requests.get(file_name)
    f = open('tiobe/' +file_name.split('/')[-1], 'wb')
    f.write(res.content)
    f.close()
```
