#### 테이블 생성 코드

```python
import sqlite3
conn = sqlite3.connect('test.db')

# 데이터를 다루기 위한 커서 생성
cursor = conn.cursor()

# 데이터베이스에 적용할 명령 준비
sql = '''
CREATE TABLE my_data (
    id INTEGER PRIMARY KEY AUTOINCREMENT, 
    a TEXT, b TEXT,
    c TEXT
)
'''

# 데이터베이스 명령 실행
cursor.execute(sql)

# 실행 결과 저장
conn.commit()

# 커서 닫기 (해제)
cursor.close() 

conn.close()
```

#### 데이터 입력 코드

#### 데이터 조회 코드

- fetchall(): 전체 조회
- fetchmany(): N개 조회
- fetchone(): 1개 조회

```python
import sqlite3
conn = sqlite3.connect('test.db')

# 데이터를 다루기 위한 커서 생성
cursor = conn.cursor()

# 데이터베이스에 적용할 명령 준비
sql = '''
SELECT id, a, b, c FROM my_data
'''

# 데이터베이스 명령 실행
cursor.execute(sql)

result = cursor.fetchall()
for row in result:
    print(row[0], row[1], row[2], row[3])

# 커서 닫기 (해제)
cursor.close() 

conn.close()
```
